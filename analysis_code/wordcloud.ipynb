```python
!pip install wordcloud
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS
dict1={}
string1=""
stopwords = STOPWORDS
for i in range(1,6):
    j='data/ar'+str(i)+'.txt'
    with open('data/ar'+str(i)+'.txt') as f:
        dict1[i]=f.read()
for i,j in dict1.items():
    string1 = string1+str(j)

wordcloud = WordCloud(stopwords=stopwords, background_color="white", max_words=1000).generate(string1)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```
